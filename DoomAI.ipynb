```python
# Import VIZDoom for the game environment
from vizdoom import *

# Import random for action sampling
import random

# Import time for sleeping
# VIZDoom runs and trains extremely quickly, therefore time will be used to artificially slow down the frame
# rate in order to actually be able properly demonstrate the AI agent playing the game to a human observer
import time

# Import numpy in order to create an identity matrix of agent actionable events and actions
import numpy as np
```

```python
# Sets up the in game environment

# Creates a game instance
game = DoomGame()

# Allows for the actual loading of the configurations of the game, such as the map

# Basic Version
#game.load_config('vizdoomGithubClone/VIZDoom/scenarios/basic.cfg')

# Defend The Center Version
#game.load_config('vizdoomGithubClone/VIZDoom/scenarios/defend_the_center.cfg')

# Deadly Corridor Version
game.load_config('vizdoomGithubClone/VIZDoom/scenarios/deadly_corridor_s1.cfg')

# Starts the game instance
game.init()
```

```python
# Retrieves the AI agent's possible actions within the in game Doom environment
# Basic & Defend The Center Versions:
# Moving to the right, to the left or to attack
# Move Right: actions[0]
# Move Left: actions[1]
# Attack: actions[2]
#
# Turns the action arrays from decimals to integers via dtype as a numpy 8 bit integer
#actions = np.identity(3, dtype = np.uint8)

# Deadly Corridor Version
actions = np.identity(7, dtype = np.uint8)
```

```python
# Inititates a loop of 10 episodes or games
episodes = 10
for episode in range(episodes):
    
    # Equivalent to rebooting & restarting the episode or game
    game.new_episode()
    
    # Checks whether the game has not ended either through the player character's death or whether the level has
    # been successfully completed
    while not game.is_episode_finished():
        # Retrieves the current state of the in game environment
        state = game.get_state()
        
        # Retrieves the actual screen image frame of the game at a point in time
        img = state.screen_buffer
        
        # Retrieves the various in game environment variables, such as current ammon count
        info = state.game_variables
        
        # Prints out the associated reward for each of the AI agent's in game actions
        # Killing an enemy: +100
        # Missing a shot: -5
        # Living: -1
        #
        # Frame Skipping: It takes time for an action to actually carry out within the game environment and
        # so this must be taken into consideration, wait or skip for 4 frames to occur before having the AI agent
        # taking a new action
        reward = game.make_action(random.choice(actions), 4)
        #print("AI Agent's Reward: ", reward)
        
        # In seconds
        time.sleep(0.02)
    print('Final Results: ', game.get_total_reward())
    
    # Initiate's a slightly longer break between episodes or games
    time.sleep(2)
```

```python
# Creates the VIZDoom OpenAI Gym Environment
class VIZDoomGym(Env):
    # Function called up when starting up a new in game environment
    
    # Basic & Defend The Center Version
    #def __init__(self, render = False):
    
    # Deadly Corrifor Version
    # Modified configuration has adjusted skill values from 1 to 5 to help the AI agent through 
    # Curriculum Learning, progressively increasing the difficulty of the environment 
    # to allow said AI agent initially learn 'good' actions early on
    def __init__(self, render = False, config = 'vizdoomGithubClone/VIZDoom/scenarios/deadly_corridor_s1.cfg'):    
        # Inherit from the Env class
        super().__init__()
        
        # Sets up the game
        self.game = DoomGame()
        
        # Basic Version
        #self.game.load_config('vizdoomGithubClone/VIZDoom/scenarios/basic.cfg')
        
        # Defend The Center Version
        self.game.load_config(config)
    
        # Rendering requires computing power, therefore by preventing rendering during the AI agent's model
        # training this can make said training much quicker by avoiding pointless computing resource usage
        if render == False:
            self.game.set_window_visible(False)
        else:
            self.game.set_window_visible(True)
        
        # Starts the game
        self.game.init()
    
        # Creates the observation and action spaces
        # Box used to contextualize the actual game's screen window environment
        # Low: Pixel Minimum
        # High: Pixel Maximum
        self.observation_space = Box(low = 0, high = 255, shape = (100, 160, 1), dtype = np.uint8)
        
        # Basic & Defend The Center Version
        #self.action_space = Discrete(3)
        
        # Deadly Corrifor Version
        self.action_space = Discrete(7)
    
        # Additional Game Variables: HEALTH, DAMAGE_TAKEN, HITCOUT, SELECTED_WEAPON_AMMO
        # Deadly Corridor Version: Included modifications for reward shaping
        # Must be extracted from the game, specifically the change in value, or deltas, between in-game frames
        self.damage_taken = 0
        self.hitcount = 0
        self.ammo = 52
    
    # How the AI agent may take a stepthin the in game environment
    def step(self, action):
        
        # Specifies possible AI agent actions as well as take steps
        
        # Basic & Defend The Center Versions
        #actions = np.identity(3) #, dtype = np.uint8)
        
        # Deadly Corrifor Version
        actions = np.identity(7)
        
        # Basic & Defend The Center Versions
        #reward = self.game.make_action(actions[action], 4)
        
        # Deadly Corrifor Version
        movement_reward = self.game.make_action(actions[action], 4)
        reward = 0
        
        # Retrieves all other required information
        # If Else logic to prevent errors whenever game has reached an end state, such as 
        # charater death or level end, returns an index of 0s to prevent said errors
        if self.game.get_state():
            state = self.game.get_state().screen_buffer
            state = self.grayscale(state)
            
            # Basic & Defend The Center Versions
            # To check the environment the returned value must be within a python dictionary format
            #ammo = self.game.get_state().game_variables[0]
            
            # Deadly Corridor Version: Reward Shaping
            game_variables = self.game.get_state().game_variables
            health, damage_taken, hitcount, ammo = game_variables
            
            # Calculates the reward deltas
            damage_taken_delta = -damage_taken + self.damage_taken
            self.damage_taken = damage_taken
            hitcount_delta = hitcount - self.hitcount
            self.hitcount = hitcount
            ammo_delta = ammo - self.ammo
            self.ammo = ammo
            
            # Weight the rewards as desires, requires testing and trial and error
            # Includes the rewards and punishments of the original configuration:
            # +1 Moving towards the end goal
            # -1 Moving away from the vest
            # -100 Dying
            # -10 Taking 1 unit of damage
            # + 200 Killing an enemy
            # -5 Using 1 unit of ammo
            reward = movement_reward + (damage_taken_delta * 10) + (hitcount_delta * 200) + (ammo_delta * 5)
            
            info = ammo
            
        else:
            state = np.zeros(self.observation_space.shape)
            info = 0
        
        # Stable Baselines dislikes lone integers, returns a dictionary whether a valid or invalid state has
        # been reached, otherwise an error is brought on during AI model training
        info = {"info":info}
        
        done = self.game.is_episode_finished()
        
        return state, reward, done, info
        
    # Defines how to render the game or its environment
    # Left empty as VIZDoom already accomplishes this for us, here for completion's sake as to what OpenAI Gym
    # typically requires
    def render():
        pass
    
    # What occurs when a new game begins
    def reset(self):
        self.game.new_episode()
        state = self.game.get_state().screen_buffer
        return self.grayscale(state)
    
    # Grayscales the game's frame as well as resize it
    def grayscale(self, observation):
        
        # cv2.cvtColor to change the color input channels and their orders
        # np.moveaxis to reorder the image array axis in order for cv2 to properly work 
        gray = cv2.cvtColor(np.moveaxis(observation, 0 ,-1), cv2.COLOR_BGR2GRAY)
        
        # Resizeing the image by scaling it down to further reduce the number of pixels the AI has to process
        # when training the agent model
        resize = cv2.resize(gray, (160, 100), interpolation = cv2.INTER_CUBIC)
        state = np.reshape(resize, (100, 160, 1))
        
        return state
    
    # Called in order to close or turn off the game window
    def close(self):
        self.game.close()
```

```python
# Import OS for file path management
import os

# Import Base Callback in order to save completed AI models
# Optional: For backup purposes every x number of steps, very large file sizes, maybe better to do manually
from stable_baselines3.common.callbacks import BaseCallback
```

```python
class TrainAndLoggingCallback(BaseCallback):
    
    def __init__(self, check_freq, save_path, verbose = 1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path
        
    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok = True)
            
    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, 'best_model_{}'.format(self.n_calls))
            self.model.save(model_path)
        
        return True
```

```python
# Create model checkpoint and logs directories

# Basic Version
#CHECKPOINT_DIR = './train/train_basic'
#LOG_DIR = './logs/log_basic'

# Defend The Center Version
#CHECKPOINT_DIR = './train/train_defend'
#LOG_DIR = './logs/log_defend'

# Deadly Corridor Version
CHECKPOINT_DIR = './train/train_corridor'
LOG_DIR = './logs/log_corridor'
```

```python
# Sets up the AI model saving callback system
callback = TrainAndLoggingCallback(check_freq = 100000, save_path = CHECKPOINT_DIR)
```

```python
# Import Proximal Policy Optimization (PPO) for chosen machine learning algorithms
from stable_baselines3 import PPO
```

```python
# Non rendered in game environment for training purposes

# Basic & Defend The Center Versions
#env = VIZDoomGym()

# Deadly Corridor Version
# Curriculum learning involves training the model in consequtively more difficult scenarios
env = VIZDoomGym(config = 'vizdoomGithubClone/VIZDoom/scenarios/deadly_corridor_s1.cfg')
```

```python
# The actual, temporary AI model that has yet to be trained, started
#
# Policy Network determines the type of neural network behind the scenes of deep learning AI models
#
# CnnPolicy (Convelutional Neural Network Polciy)chosen in this instance as it is very quick in terms of 
# learning with images, in our case frames from the video game Super Mario Brothers, MlpPolicy for tabular data
#
# env: The prepocessed data
# verbose: Obtain a ton of information once the training begins
# tensorboard_log: Information as to how the training is performing as the model runs
# learning_rate: Critical, learn to quickly and the AI may converge to a poor model, 
#                train to slowly and it might take years
# n_steps: How may frames per game before updating the neural network,so up to x sets of observations, actions, 
#          log probailities & values will be stored in the buffer for one iteration
#          The batch size being passed to the actor and critic of the neural network, or how many time frames
#          being passed in total as part of one training run, so the larger the value the more learning per epoch
#          Basic Version: 2048
#          Defend The Center Version: 4096
#          Deadly Corridor:
#          CnnPolicy requires a batch size as a function of 64
#model = PPO('CnnPolicy', env, verbose = 1, tensorboard_log = LOG_DIR, learning_rate = 0.0001, n_steps = 4096)
model = PPO('CnnPolicy', env, verbose = 1, tensorboard_log = LOG_DIR, learning_rate = 0.00001, n_steps = 8192,
            clip_range = 0.1, gamma = 0.95, gae_lambda = 0.9)
```

```python
# Trains the AI model and where it begins to actually learn
#
# total_timesteps: How long via how many frames the model should learn
model.learn(total_timesteps = 100000, callback = callback) # If automatic callback not desired
```

```python
# Curriculum Learning Setup

# Load the previously trained model on an easier difficulty
model.load('./train/train_corridor/best_model_100000')

# Pass pre-trained model into more dificult environment for more strenous learning
env = VIZDoomGym(config = 'vizdoomGithubClone/VIZDoom/scenarios/deadly_corridor_s2.cfg')
model.set_env(env)
model.learn(total_timesteps = 40000, callback = callback)
```
